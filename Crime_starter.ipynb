```python
#Import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#bring in dataframe

csv='C:/Users/Daniel/Downloads/arrest-data-from-2010-to-present_clean.csv'
df = pd.read_csv(csv,low_memory=False)
df
```

```python
#Set variables for the areas and the crimes

area_names=set(df['Area Name'])
crimes=set(df['Charge Group Description'])
```

```python
#Group by Area name to have distinct areas

df.groupby('Area Name').head()
```

```python
#Dataframe for area
df_area=df[['Area Name','Reporting District','Address','Location','Zip Codes','Charge Group Description','Charge Group Code']]
df_area
```

```python
#create groups that sort by area name and return the amount of crimes of each type committed
num_crimes_in_district=df_area.groupby(['Area Name'])["Charge Group Description"].value_counts()
num_crimes_in_district
```

```python
#Create new dataframe of the above groupby and add in a column 'Number of crimes'
plot_df  = pd.DataFrame(num_crimes_in_district)
plot_df.columns = ['Number of Crimes']
plot_df
```

```python
#Create list of different types of crimes that fall under a similar category
print (crimes)
heavy_crimes =['Aggravated Assault','Homicide','Rape','Robbery',]
bad_for_cars=['Vehicle Theft','Driving Under Influence','Moving Traffic Violations']
white_collar_crimes =['Fraud/Embezzlement','Forgery/Counterfeit','Gambling']
#Create a dataframe for each of the different categories by resetting the indexes to 
#then search for only the crimes that fall in the list specified
hc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(heavy_crimes)]
bfc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(bad_for_cars)] 
wcc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(white_collar_crimes)] 
```

```python
#Update the dataframes with the groups back in place
hc_locations_df = hc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
wcc_locations_df= wcc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
bfc_locations_df = bfc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
```

```python
#reworking groups in df_area to make a dataframe to find the top crime commited in each district
biggest_crime_in_district=df_area.groupby(["Charge Group Description"])["Area Name"].value_counts()
bcid_df=pd.DataFrame(biggest_crime_in_district)
bcid_df.columns=['Number of Crimes']
top_crime_list=[]
for crime in crimes:

    crime_variable=bcid_df.loc[[(crime)]]
    top_crime_list.append(crime_variable.iloc[[0]])
tc_df = pd.concat(top_crime_list)

tc_df.reset_index(inplace=True)

tc_df.set_index(['Charge Group Description'],inplace=True)
```

```python
#Creating a graph for top crime by area
##same legend making process as above
mycolors = ['red','blue','yellow','orange','purple']
myitems = tc_df['Area Name'].unique()
colors = {}
mycolors_legend = []

for i in range(len(mycolors)):
    colors[myitems[i]] = mycolors[i]
    mycolors_legend.append(myitems[i])
myco = []
for i in range(len(tc_df)):
    myco.append(colors[tc_df['Area Name'][i]])

height = list(tc_df['Number of Crimes'])
bars = tuple(tc_df.index)#['Charge Group Description'])
y_pos = np.arange(len(bars))

plt.bar(y_pos, height, color=tuple(myco))
plt.xticks(y_pos, bars,rotation=90)
ke = []
custom_lines = []
for key,val in colors.items():
    custom_lines.append(Line2D([0], [0], color=val, lw=4))
    ke.append(key)
plt.legend(custom_lines,ke,loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.title('Location where most arrests occur', color='black')

plt.savefig('total_crime_by_area.png',bbox_inches='tight')
plt.show()

```
