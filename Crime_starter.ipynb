```python
plot_df  = pd.DataFrame(num_crimes_in_district)
```

```python
#safest Spaces
print (crimes)
heavy_crimes =['Aggravated Assault','Homicide','Rape','Robbery',]
bad_for_cars=['Vehicle Theft','Driving Under Influence','Moving Traffic Violations']
white_collar_crimes =['Fraud/Embezzlement','Forgery/Counterfeit','Gambling']
hc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(heavy_crimes)]
bfc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(bad_for_cars)] 
wcc_locations_df = plot_df.reset_index(inplace=False)[plot_df.reset_index(inplace=False)['Charge Group Description'].isin(white_collar_crimes)] 
```

```python
hc_locations_df = hc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
wcc_locations_df= wcc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
bfc_locations_df = bfc_locations_df.groupby(['Area Name','Charge Group Description']).sum()
```
