What time of year are arrests in LA more common?
What time of day are arrests in LA more common?


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("arrest-data-from-2010-to-present_clean.csv")
```

```python
#arrests by year
data['year'] = pd.Series(data['Arrest Date'].str[:4],index=data.index)
crimeByYear = data['year'].value_counts(sort=False).sort_index()
crimeByYear.plot(kind = 'line')
plt.title("Arrests By Year")
plt.xlabel("Year")
plt.ylabel("Number Of Arrests")
```

```python
#graph arrests by time / a day
data['time'] = pd.Series(data['Time'],index=data.index)
crimeByTime = data['time'].value_counts(sort=False).sort_index()
crimeByTime.plot(kind = 'line')
plt.title("Arrests By Time")
plt.xlabel("Time Per Day")
plt.ylabel("Number Of Arrests")
```
